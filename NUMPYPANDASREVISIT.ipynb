```python
import numpy as np

a = np.array([
    [1, 2, 3],
    [2, 3, 4],
    [4, 2, 1]
])

print("First matrix is:")
print(a)
print("*********************************")

b = np.array([
    [2, 8, 0],
    [6, 3, 2],
    [4, 5, 7]
])

print("Second matrix is:")
print(b)
print("*********************************")

print("Addition of both matrices is")
print(a + b)
print("*********************************")

print("Subtraction of both matrices is")
print(a - b)
print("*********************************")

print("Multiplication of both matrices is")
print(np.dot(a, b))
print("*********************************")

print("Inversion of first matrix is")
z = np.linalg.inv(a)
print(z)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output
import numpy as np

adani_ports = pd.read_csv("ADANIPORTS.csv")
asian_paint = pd.read_csv("ASIANPAINT.csv")
axis_bank = pd.read_csv("AXISBANK.csv")
tcs = pd.read_csv("TCS.csv")
tata_motors = pd.read_csv("TATAMOTORS.csv")
tata_steel = pd.read_csv("TATASTEEL.csv")
titan = pd.read_csv("TITAN.csv")
wipro = pd.read_csv("WIPRO.csv")
zeel = pd.read_csv("ZEEL.csv")
ntpc = pd.read_csv("NTPC.csv")
ongc = pd.read_csv("ONGC.csv")
kotak_bank = pd.read_csv("KOTAKBANK.csv")
jsw_steel = pd.read_csv("JSWSTEEL.csv")
maruti = pd.read_csv("MARUTI.csv")
reliance = pd.read_csv("RELIANCE.csv")
sunpharma = pd.read_csv("SUNPHARMA.csv")
bpcl = pd.read_csv("BPCL.csv")
britannia = pd.read_csv("BRITANNIA.csv")
bajajfinsv = pd.read_csv("BAJAJFINSV.csv")
bajfinance = pd.read_csv("BAJFINANCE.csv")
hdfc_bank = pd.read_csv("HDFCBANK.csv")
icici_bank = pd.read_csv("ICICIBANK.csv")
techm = pd.read_csv("TECHM.csv")
```

```python

for df in [tata_motors, tata_steel, tcs, titan, wipro, zeel, ntpc, ongc, kotak_bank, jsw_steel, maruti, reliance, sunpharma, bpcl, britannia, bajajfinsv,
           bajfinance, hdfc_bank, icici_bank, techm, adani_ports, asian_paint]:
    df.drop(['Trades', 'Deliverable Volume', '%Deliverble'], axis=1, inplace=True)


for df in [tata_motors, tata_steel, tcs, titan, wipro, zeel, ntpc, ongc, kotak_bank, jsw_steel, maruti, reliance, sunpharma, bpcl, britannia, bajajfinsv,
           bajfinance, hdfc_bank, icici_bank, techm, adani_ports, asian_paint]:
    df['Date'] = pd.to_datetime(df['Date'])


stocks_dict = {
    'Adani Ports': adani_ports,
    'Asian Paint': asian_paint,
    'Axis Bank': axis_bank,
    'TCS': tcs,
    'Tata Motors': tata_motors,
    'Tata Steel': tata_steel,
    'Titan': titan,
    'Wipro': wipro,
    'ZEEL': zeel,
    'NTPC': ntpc,
    'ONGC': ongc,
    'Kotak Bank': kotak_bank,
    'JSW Steel': jsw_steel,
    'Maruti': maruti,
    'Reliance': reliance,
    'Sun Pharma': sunpharma,
    'BPCL': bpcl,
    'Britannia': britannia,
    'Bajaj Finserv': bajajfinsv,
    'Bajaj Finance': bajfinance,
    'HDFC Bank': hdfc_bank,
    'ICICI Bank': icici_bank,
    'Tech Mahindra': techm
}


```

```python
def display_statistics(stock_name):
    stock = stocks_dict[stock_name]
    stats = stock['Close'].describe()
    return f"Statistics for {stock_name}:\n{stats}"
```

```python

def calculate_statistics(stock_name):
    stock = stocks_dict[stock_name]
    mean = np.mean(stock['Close'])
    variance = np.var(stock['Close'])
    std_dev = np.std(stock['Close'])
    return (f"Mean: {mean:.2f}\n"
            f"Variance: {variance:.2f}\n"
            f"Standard Deviation: {std_dev:.2f}")
```

```python

def correlation_between_stocks(stock1_name, stock2_name):
    stock1 = stocks_dict[stock1_name]['Close']
    stock2 = stocks_dict[stock2_name]['Close']
    correlation = stock1.corr(stock2)
    return f"Correlation between {stock1_name} and {stock2_name}: {correlation:.2f}"
```

```python

def plot_stock_prices(stock_name):
    stock = stocks_dict[stock_name]
    plt.figure(figsize=(10, 5))
    plt.plot(stock['Date'], stock['Close'], label=stock_name)
    plt.title(f'Stock Prices Over Time: {stock_name}')
    plt.xlabel('Date')
    plt.ylabel('Closing Price')
    plt.legend()
    plt.grid()
    plt.show()

def compare_stocks(stock1_name, stock2_name):
    stock1 = stocks_dict[stock1_name]
    stock2 = stocks_dict[stock2_name]

    plt.figure(figsize=(10, 5))
    plt.plot(stock1['Date'], stock1['Close'], label=stock1_name, color='blue')
    plt.plot(stock2['Date'], stock2['Close'], label=stock2_name, color='orange')

    plt.title(f'Comparison of Stock Prices: {stock1_name} vs {stock2_name}')
    plt.xlabel('Date')
    plt.ylabel('Closing Price')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
stock_options = list(stocks_dict.keys())

stock1_dropdown = widgets.Dropdown(options=stock_options, description='Stock 1:')
stock2_dropdown = widgets.Dropdown(options=stock_options, description='Stock 2:')
action_dropdown = widgets.RadioButtons(options=['stats', 'plot', 'correlation'], description='Action:')
run_button = widgets.Button(description='Run Analysis')

output_area = widgets.Output()
```

```python

def on_button_click(b):
    output_area.clear_output()
    stock1 = stock1_dropdown.value
    stock2 = stock2_dropdown.value
    action = action_dropdown.value

    with output_area:
        if stock1 == stock2 and action != 'plot':
            print("Please select two different stocks for this action.")
            return
        if action == 'stats':
            print(display_statistics(stock1))
            print(calculate_statistics(stock1))
            print("\n")
            print(display_statistics(stock2))
            print(calculate_statistics(stock2))
        elif action == 'plot':
            if stock1 == stock2:
                plot_stock_prices(stock1)
            else:
                compare_stocks(stock1, stock2)
        elif action == 'correlation':
            print(correlation_between_stocks(stock1, stock2))
        else:
            print("Unknown action")

run_button.on_click(on_button_click)

display(stock1_dropdown, stock2_dropdown, action_dropdown, run_button, output_area)
```

```python
sample_stock = stocks_dict['TCS']

close_prices = sample_stock['Close'].to_numpy()

median_price = np.median(close_prices)
print("Median Close Price:", median_price)

max_price = np.max(close_prices)
min_price = np.min(close_prices)
print("Max Price:", max_price)
print("Min Price:", min_price)

cumsum_prices = np.cumsum(close_prices)
print("Cumulative Sum (first 5):", cumsum_prices[:5])

sorted_prices = np.sort(close_prices)
print("Sorted Close Prices (last 5):", sorted_prices[-5:])

noise = np.random.normal(0, 1, len(close_prices))
noisy_close = close_prices + noise
print("Noisy Prices Sample:", noisy_close[:5])



sample_stock['RollingAvg_7'] = sample_stock['Close'].rolling(window=7).mean()
print("7-day Rolling Avg (last 5):\n", sample_stock['RollingAvg_7'].tail())

sample_stock['Daily_Return'] = sample_stock['Close'].pct_change() * 100
print("Daily Return % (first 5):\n", sample_stock['Daily_Return'].head())

sample_stock['Close_Rank'] = sample_stock['Close'].rank()
print("Ranking by Close (last 5):\n", sample_stock[['Date', 'Close', 'Close_Rank']].tail())

sample_stock.set_index('Date', inplace=True)
print("Date set as index. DataFrame head:\n", sample_stock.head())

monthly_avg = sample_stock['Close'].resample('M').mean()
print("Monthly Avg Close:\n", monthly_avg.tail())
```
